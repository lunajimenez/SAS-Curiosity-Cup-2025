```python
import os
import pandas as pd
import matplotlib.pyplot as plt

carpeta_data = "./data" #Ruta de la carpeta donde están los CSV
archivos_csv = [f for f in os.listdir(carpeta_data) if f.endswith(".csv")] #Archivos CSV listados en la carpeta

print(f"Se encontraron {len(archivos_csv)} archivos CSV en la carpeta 'data'.")
print("\nLista de archivos:")

for archivo in archivos_csv:
    print(f"- {archivo}")

# Cargar y mostrar las primeras filas de cada CSV
for archivo in archivos_csv:
    ruta_archivo = os.path.join(carpeta_data, archivo)
    print(f"\nMostrando datos de: {archivo}")
    try:
        df = pd.read_csv(ruta_archivo, encoding='latin1')  # Intentamos leer con encoding 'latin1'
    except UnicodeDecodeError:
        df = pd.read_csv(ruta_archivo, encoding='utf-8')  # Si falla, intentamos con 'utf-8'
    print(df.head())
    print("\nInformación general:")
    print(df.info())
    print("-"*80)
    
    # Si el archivo tiene columnas numéricas, mostrar un histograma de la primera columna numérica
    #columnas_numericas = df.select_dtypes(include=['number']).columns
    #if len(columnas_numericas) > 0:
        #df[columnas_numericas[0]].hist(bins=20)
        #plt.title(f"Histograma de {columnas_numericas[0]} en {archivo}")
        #plt.show()
```
